```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
no_brokers = requests.get('https://www.nobroker.in/property/sale/bangalore/Electronic%20City?type=BHK4&searchParam=W3sibGF0IjoxMi44NDUyMTQ1LCJsb24iOjc3LjY2MDE2OTUsInBsYWNlSWQiOiJDaElKdy1GUWQ0cHNyanNSSGZkYXpnXzhYRW8iLCJwbGFjZU5hbWUiOiJFbGVjdHJvbmljIENpdHkifV0=&propertyAge=0&radius=2.0')
no_brokers
```

```python
no_brokers.content
```

```python
soup = BeautifulSoup(no_brokers.content)
soup
```

```python
print(soup.prettify)
```

```python
first_flat = soup.find('h2', class_="heading-6 font-semi-bold nb__1AShY")
first_flat
```

```python
first_flat.text
```

```python
all_flats = soup.find_all('h2',class_="heading-6 font-semi-bold nb__1AShY")
all_flats
```

```python
main_flats = []
for i in all_flats:
    main_flats.append(i.text)
main_flats    
```

```python
address = soup.find('div', class_="nb__2CMjv")
address
```

```python
address.text
```

```python
all_address = soup.find_all('div', class_="nb__2CMjv")
all_address
```

```python
main_address = []
for i in all_address:
    main_address.append(i.text)
main_address    
```

```python
builtup_area = soup.find('div',class_="nb__3oNyC")
builtup_area
```

```python
builtup_area.text
```

```python
all_builtup_area = soup.find_all('div',class_="nb__3oNyC")
all_builtup_area
```

```python
main_builtup_area=[]
for i in all_builtup_area:
    main_builtup_area.append(i.text)
main_builtup_area    
```

```python
estimated_emi = soup.find_all('div',class_="font-semi-bold heading-6")
estimated_emi 
```

```python
main_emi=[]
for i in estimated_emi:
    main_emi.append(i.text.replace('\n',''))
main_emi
```

```python
del main_emi[::3]
main_emi
```

```python
final_estimated_emi = main_emi[::2]
```

```python
final_estimated_emi 
```

```python
price = soup.find_all('div',class_="font-semi-bold heading-6")
price
```

```python
main_price=[]
for i in price:
    main_price.append(i.text.replace('\n',''))
main_price    
```

```python
del main_price[::3]
main_price
```

```python
final_price = main_price[::-2]
```

```python
final_price
```

```python
print(len(main_flats),len(main_address),len(main_builtup_area),len(final_estimated_emi),len(final_price))
```

```python
final_nobrokers = pd.DataFrame({})
final_nobrokers['House Title'] = main_flats
final_nobrokers['Location'] = main_address 
final_nobrokers['Area'] = main_builtup_area 
final_nobrokers['Emi'] = final_estimated_emi
final_nobrokers['Prices'] = final_price
```
